```python
import pandas as pd
```

```python
games["date"] = pd.to_datetime(games["date"]) # converting the date col to datetime type
```

```python
games["opp_code"] = games["opp"].astype("category").cat.codes # converting opponents to an opp code
```

```python
import numpy as np
```

```python
games["target"] = np.where(games["w/l"].str.startswith("W"), 1, 0) # converting wins/losses to usable data
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=200, min_samples_split=15, random_state=1)
```

```python
train = games[games["date"] < '2023-05-01'] # train on old data
```

```python
test = games[games["date"] > '2023-05-01'] # test training based model on newer data
```

```python
predictors = ["venue_code", "opp_code", "day_code"] # initial predictors for model
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
from sklearn.metrics import precision_score
```

```python
grouped_games = games.groupby("team")
```

```python
def rolling_averages(group, cols, new_cols): #calculates rolling average of cols
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["tm", "opp.1", "fg%", "fg%.1", "3p%", "3p%.1", "ft%", "ft%.1", "stl", "stl.1"] # new cols for predictors
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
games_rolling = games.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols), include_groups=False)
```

```python
games_rolling
```

```python
games_rolling = games_rolling.reset_index()
```

```python
games_rolling
```

```python
games_rolling.index = range(games_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2023-05-01']
    test = data[data["date"] > '2023-05-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted = preds), index = test.index)
    precision = precision_score(test["target"], preds)
    accuracy = accuracy_score(test["target"], preds)
    return combined, precision, accuracy
```

```python
combined, precision, accuracy = make_predictions(games_rolling, predictors + new_cols)
```
